```python
# importing Libaries
import numpy as np
import cv2

# Reading Image
color_image = cv2.imread(r'E:\4Balls.jpg')

#resise image
color_image = cv2.resize(color_image, (300,300))

# Convert RGB image to HSV image
hsv_image = cv2.cvtColor(color_image, cv2.COLOR_RGB2HSV)

# define lower and Upper range for Red, Green, Blue color
lower_red = (0,50,50)
upper_red = (180,255,255)

lower_green = (35,100,100)
upper_green = (86,255,255)

lower_blue = (100,150,0)
upper_blue = (140,255,255)

#Applying Mask to image
red_mask = cv2.inRange(hsv_image, lower_red, upper_red)
green_mask = cv2.inRange(hsv_image, lower_green, upper_green)
blue_mask = cv2.inRange(hsv_image, lower_blue, upper_blue)
red_image = cv2.bitwise_and(color_image, color_image, mask= red_mask)
green_image = cv2.bitwise_and(color_image, color_image, mask= green_mask)
blue_image = cv2.bitwise_and(color_image, color_image, mask= blue_mask)

#Displying image
cv2.imshow('Original image', color_image)
cv2.imshow('Red Color Detected', red_image)
cv2.imshow('Green Color Detected', green_image)
cv2.imshow('Blue Color Detected', blue_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Another Code
import numpy as np
import cv2

# Reading Image
#color_image = cv2.imread(r'E:\\4Balls.jpg')
#color_image = cv2.imread(r'E:\\flowers.jpeg')
color_image = cv2.imread(r'E:\\pencils.jpg')

# Resize image
color_image = cv2.resize(color_image, (300, 300))

# Convert BGR image to HSV image
hsv_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2HSV)

# Define lower and upper range for Red, Green, Blue color
lower_red1 = (0, 120, 70)
upper_red1 = (10, 255, 255)
lower_red2 = (170, 120, 70)
upper_red2 = (180, 255, 255)

lower_green = (35, 100, 100)
upper_green = (85, 255, 255)

lower_blue = (94, 80, 2)
upper_blue = (126, 255, 255)

# Applying Mask to image
red_mask1 = cv2.inRange(hsv_image, lower_red1, upper_red1)
red_mask2 = cv2.inRange(hsv_image, lower_red2, upper_red2)
red_mask = red_mask1 + red_mask2

green_mask = cv2.inRange(hsv_image, lower_green, upper_green)
blue_mask = cv2.inRange(hsv_image, lower_blue, upper_blue)

# Masked images
red_image = cv2.bitwise_and(color_image, color_image, mask=red_mask)
green_image = cv2.bitwise_and(color_image, color_image, mask=green_mask)
blue_image = cv2.bitwise_and(color_image, color_image, mask=blue_mask)

# Displaying images
cv2.imshow('Original Image', color_image)
cv2.imshow('Red Color Detected', red_image)
cv2.imshow('Green Color Detected', green_image)
cv2.imshow('Blue Color Detected', blue_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
